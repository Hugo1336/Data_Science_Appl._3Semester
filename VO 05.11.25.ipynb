```python
import os

# ensure output directory exists
os.makedirs("quotes", exist_ok=True)

base_url = "https://quotes.toscrape.com/page/"

for page in range(1, 10):
    # Eigentliche Seiten URL zusammenbauen
    url = base_url + str(page)
    print(url)
    # Seite abfragen
    response = get(url)
    # Suchkartei erzeugen
    soup = BeautifulSoup(
        response.content,
        "html.parser"
    )
#alle zitate boxen in html finden
    quotes = soup.find_all("div", attrs={"class":"quote"})
    for index, quote in enumerate(quotes):
        # include page in filename to avoid collisions

        statement = quote.find_next("span", attrs={"class":"text"})
        # Nur den reinen Zitattext extrahieren (ohne Autor / Tags)
        quote_text = statement.get_text(strip=True).strip('“”"')
        filename = f"quotes/quote_{page}_{index}.txt"
        with open(filename, "w", encoding="utf-8") as file:
            # write only the cleaned quote text
            file.write(quote_text)

    # be polite to the server
    sleep(3)
```
