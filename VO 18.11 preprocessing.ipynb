```python
from nltk.corpus import stopwords
from text_magic.functions import clean_strip
```

```python
def clean_stopwords(text:str, language:str='english'):
    words = text.split(" ")
    clean_text = []
    for word in words:
        if word not in stopwords.words(language):
            clean_text.append(word)

    return " ".join(clean_text)

```

```python
text = "Hallo Welt wie geht es dir"
clean_stopwords(text, language='german')
```
